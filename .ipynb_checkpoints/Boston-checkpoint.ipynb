## Split Datasetm

    

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(boston_df[x_feature], boston_df.medv, test_size = 0.2)
```

```python
x_train
```

```python
x_train.columns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
lreg.fit(x_train, y_train)
```

```python

lreg.intercept_
```

```python
y_pred = lreg.predict( x_test )
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

R.sqr = (Ma - Me) / Ma       

Ma = Model derive error
Mm = Model making error



## Build NN model

### Normalize Data

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation 
```

```python
import keras
```

```python
import keras.backend as k
```

```python
model = Sequential()
```

```python
X_train_scaled = x_train.shape
```

```python
X_train_scaled
```

```python

model.add(Dense(32, input_shape = (X_train_scaled[1],)))
model.add(Dense(16))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('relu'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer='rmsprop', 
              metrics=['accuracy'])
```

```python
history = model.fit
```

```python
EPOCHS = 100
BATCH_SIZE = 25
```

```python
%%time 
# fit model
history = model.fit(
    x_train, 
    y_train,  # prepared data
    epochs = EPOCHS,
    validation_data = (x_test, y_test),
    shuffle = True,
    verbose=0,
)
```

```python
history.history
```

```python
val_mse = history.history['val_mean_squared_error']
```

```python
x_test_scaled = x_test.shape
```

```python
x_test_scaled
```

```python
y_pred = model.predict(x_test_scaled)
```
